# Importing the Required Data 

```python
data=pd.read_excel("All_Schemes_Data_Broken_Into_Mom_And_Baby.xlsx", sheet_name="Mom_Data",index_col=[0,1])
```

```python
data_13 = pd.read_excel("MATERNITY_MEM_2013.xlsx")
data_14 = pd.read_excel("MATERNITY_MEM_2014.xlsx")
data_15 = pd.read_excel("MATERNITY_MEM_2015.xlsx")
data_16 = pd.read_excel("MATERNITY_MEM_2016.xlsx")
data_17 = pd.read_excel("MATERNITY_MEM_2017.xlsx")
data_18 = pd.read_excel("MATERNITY_MEM_2018.xlsx")
data_19 = pd.read_excel("MATERNITY_MEM_2019.xlsx")
data_20 = pd.read_excel("MATERNITY_MEM_2020 ( Jan-Mar).xlsx")
```

```python
scheme_B=pd.read_excel("B_MATERNITY_EXPOSURE.xlsx")
scheme_C=pd.read_excel("C_MATERNITY_MEM_EXPOSURE.xlsx")
```

```python
data_exp.columns=data_exp.columns.str.title()
```

```python
exp_13=round((data_exp_13["Exposure"].sum())/12,2)
exp_14=round((data_exp_14["Exposure"].sum())/12,2)
exp_15=round((data_exp_15["Exposure"].sum())/12,2)
exp_16=round((data_exp_16["Exposure"].sum())/12,2)
exp_17=round((data_exp_17["Exposure"].sum())/12,2)
exp_18=round((data_exp_18["Exposure"].sum())/12,2)
exp_19=round((data_exp_19["Exposure"].sum())/12,2)
exp_20=round((data_exp_20["Exposure"].sum())/12,2)
```
